```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(['구매 품목명','Age Group','Purchase Category','Payment Type'],axis=1)
```

```python
a = 0
for row in df['고객ID'] :
  b = df.loc[a,'고객ID']
  df.loc[a,'고객ID'] = b[0:4] + '***' + b[-1]
  a += 1
df
```

```python
a = 0
for row in df['결제 유형'] :
  if (row == '현금') :
    df.loc[a,'결제카드사 종류'] = '현금'
    df.loc[a,'일시불/할부 여부'] = '일시불'
  a += 1
df
```

```python
a = 0
for row in df['상세 집주소'] :
  b = df.loc[a,'상세 집주소']
  c = b.split(' ')
  df.loc[a,'상세 집주소'] = c[0]
  a += 1
df
```

```python
a = 0
for row in df['고객명'] :
  b = df.loc[a,'고객명']
  df.loc[a,'고객명'] = b[0]+'**'
  a += 1
df
```

```python
a = 0
for row in df['나이'] :
  if (row<30) :
    df.loc[a,'나이'] = '20대'
  elif (row<40) :
    df.loc[a,'나이'] = '30대'
  elif (row<50) :
    df.loc[a,'나이'] = '40대'
  elif (row<60) :
    df.loc[a,'나이'] = '50대'
  else :
    df.loc[a,'나이'] = '60대'
  a += 1
df
```

```python
a = 0
for row in df['핸드폰번호'] :
  b = df.loc[a,'핸드폰번호']
  df.loc[a,'핸드폰번호'] = b[0:5]+'***-****'
  a += 1
df
```

```python
a = 0
for row in df['결제 금액'] :
  b = df.loc[a,'결제 금액']
  c = round(b,-4)
  df.loc[a,'결제 금액'] = c
  a += 1
df
```

```python
a = 0
for row in df['결제 시간'] :
  b = df.loc[a,'결제 시간']
  df.loc[a,'결제 시간'] = b[0:7]
  a += 1
df
```

```python
a = 0
for row in df['상세 집주소'] :
  b = df.loc[a,'상세 집주소']
  c = b.split(' ')
  df.loc[a,'상세 집주소'] = c[0]
  a += 1
df
```
